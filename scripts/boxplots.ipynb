```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
fn_server_resp = 'server-response-times.json'
fn_legacy_js = 'legacy-javascript.json'
fn_redirects = 'redirects.json'
```

```python
def extract_data(filename):
    overallSavingsMs = []
    score = []
    numericValue = []
    
    with open(filename) as j_file:
        this_dict = json.load(j_file)
        for url in this_dict:
            try:
                overallSavingsMs.append(this_dict[url]['overallSavingsMs'])
                score.append(this_dict[url]['score'])
                numericValue.append(this_dict[url]['numericValue'])
            except:
                pass
            
    return (overallSavingsMs, score, numericValue)
                
```

**Server Response times:** 

```python
(overallSavingsMs, score, numericValue) = extract_data(fn_server_resp)
```

```python
df = pd.DataFrame(overallSavingsMs)
df.quantile([0.25,0.5,0.75])
```

```python
ax = df.boxplot()
plt.xlabel('overallSavings (ms)')
plt.title("Overall Savings in Server Response times")
```

```python
ax = sns.violinplot(data=df)
plt.xlabel('overallSavings (ms)')
plt.title("Overall Savings in Server Response times")
```

```python
df = pd.DataFrame(score)
df.quantile([0.25,0.5,0.75])
```

```python
ax = df.boxplot()
plt.xlabel('Score')
plt.title("Scores in Server Response times")
```

```python
ax = sns.violinplot(data=df)
plt.xlabel('Score')
plt.title("Scores in Server Response times")
```

```python
df = pd.DataFrame(numericValue)
df.quantile([0.25,0.5,0.75])
```

```python
print('Mean:', df.mean()[0], '\nStd Dev:', df.std()[0])
# print('Number of websites with greater than 0ms potential savings:', np.sum(df[0] > 0), 'out of 50')
```

```python
ax = df.boxplot()
plt.xlabel('Response time actual (ms)')
plt.title("Response times in Server Response times")
```

```python
ax = sns.violinplot(data=df)
plt.xlabel('Response time actual (ms)')
plt.title("Response times in Server Response times")
```
